```python
import scipy
```

```python
from scipy.stats import binom

n = 10  # number of trials
p = 0.04  # probability of success on a single trial
x = 3  # number of successes
cum_probability = binom.cdf(x, n, p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

x_ = np.arange(0, n+1)
pmf = binom.pmf(x_, n, p)

plt.bar(x_, pmf)
plt.xlabel('N Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=10, p=0.04)')
plt.show()
```

```python
from scipy.stats import poisson

k = 12
lambda_ = 19

pmf = poisson.pmf(k, lambda_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 30)
pmf_dist = poisson.pmf(x, lambda_)

plt.bar(x, pmf_dist)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (lambda=19)')
plt.show()
```

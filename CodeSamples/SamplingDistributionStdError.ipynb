# Sampling Distribution of a Statistic

A **sampling distribution** refers to the probability distribution (on X axis is an event, on Y axis is probability of the event occurring) of a statistic (like a sample mean, proportion, variance, etc.) obtained from a large number of samples drawn from the same population. It describes how the statistic (e.g., proportion, sample mean) varies from sample to sample.

### Math formulas:

$$
\mu_{\hat{p}} = \mu
$$

$$
E(\bar{X}) = \mu
$$

$$
\sigma_{\hat{p}} = \frac{\sigma} {\sqrt{n}}
$$

$$
D^2(\bar{X}) = \frac{\sigma^2} {n}
$$

### Example: Sampling Distribution for Proportion of Students Using MacBooks

Assume:
- The true **proportion** of students using MacBooks in the population is $p = 0.40$ (40%).
- We take random **samples** of $n = 10$ students.
- We will simulate this process multiple times (letâ€™s say A LOT OF times) and calculate the proportion of people using MacBooks in each sample, we call it $\hat{p}$. 
- After collecting the results of all the samples, we will plot them on a **histogram** to visualize the **sampling distribution**.

### Steps:

1. At the university ask $n = 10$ people what computer they are using
    - Calculate the proportion of people that are using MacBooks to those who don't
    - Record the proportion $\hat{p}$ for the histogram 
2. Repeat the process:
    - Repeat the process of taking $n = 10$ people at random A LOT (ex. 1 000 of) times
    - For each calculate and record sample proportion $\hat{p}$
3. Visualisation of the data
    - Plot the histogram or create a table with proportion and probability of proportion occurring


### Example of Simulated Data:

| No. People w/ MB | No. such samples |
|------------------|------------------|
| 1                | 29               |
| 2                | 113              |
| 3                | 219              |
| 4                | 269              |
| 5                | 215              |
| 6                | 102              |
| 7                | 42               |
| 8                | 10               |
| 9                | 1                |
| 10               | 0                |

### Visualise the data

```python
import matplotlib.pyplot as plt

# Data from the trials
proportion = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
probabilities = [29, 113, 219, 269, 215, 102, 42, 10, 1, 0]

# Plotting the distribution
plt.figure(figsize=(10, 6))
plt.bar(proportion, probabilities, color='blue', alpha=0.7)
plt.xlabel('Number of students in a sample with MacBooks')
plt.ylabel('Number of sample groups with that value')
plt.title('Visualisation of sampling distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Compare with the theory

Now that we've done an experiment we may compare your results with ones that we would have gotten from the formula

1. Calculate the mean using the formula
    $$
     \mu_{\hat{p}} = \mu = 0.4
    $$

2. Calculate the mean from the samples
    $$
    \mu_{\hat{p}} = \sum \hat{p} P(\hat{p}) = 0.1 * 0.029 + 0.2 * 0.113 + ... + 0.9 * 0.001 + 1 * 0 = 0.406
    $$


```python
import numpy as np

prop_a = np.array(proportion)  # Convert to np array
prop_a = prop_a / 10  # convert from no people in group to proportion

prob_a = np.array(probabilities)  # Convert to np array
prob_a = prob_a / 1000  # Convert from number of occurrences to probability

mean = np.sum(prop_a * prob_a)
print(f"Mean: {mean}")
```
